```python
# import the necessary packages
import numpy as np
import imutils
import cv2


from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("Teaching Gels/Gel1_2019-05-19_Hts105_3_8bit.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
plt.imshow(gray)
```

```python
rows,cols,ch = img.shape

pts1 = np.float32([[60,700],[60,200], [1025,200], [1100,680]])
pts2 = np.float32([[0,500],[0,0], [800,0], [800,500]])

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(800,500))

plt.imshow(img),plt.title('Input')
```

```python
plt.imshow(dst),plt.title('Output')
```
